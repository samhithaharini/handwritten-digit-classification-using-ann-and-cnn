# Experiment: Handwritten Digit Classification using ANN and CNN

## Objective
1. Implement an Artificial Neural Network (ANN)
2. Implement a Convolutional Neural Network (CNN)
3. Compare their performance


## Step 1: Import Libraries

👉 Question:
- Which libraries are required for building neural networks?
- Which library provides the MNIST dataset?
- Why do we use NumPy in machine learning?


#### Libraries required:
- numpy → numerical operations
- matplotlib → visualization
- tensorflow / keras → building neural networks

#### MNIST dataset library:
- keras.datasets.mnist

#### Why NumPy?
- Efficient numerical computation
- Handles arrays/matrices
- Speeds up mathematical operations

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers
```

## Step 2: Load Dataset

👉 Question:
- What is the MNIST dataset?
- How many training and testing samples are available?
- What is the shape of one image?


#### MNIST dataset:
Collection of 70,000 handwritten digit images (0–9)
#### Samples:
Training: 60,000
Testing: 10,000
#### Shape of one image:
(28, 28) pixels (grayscale)

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

## Step 3: Data Preprocessing

👉 Question:
- Why do we normalize pixel values?
- What is the range of pixel values before and after normalization?
- How does normalization help in training?


#### Why normalize?
- To scale data for faster and stable training

#### Pixel range:
- Before: 0–255
- After: 0–1

#### How it helps:
- Improves convergence
- Prevents large gradients

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

# PART 1: ANN MODEL


## Step 4: Flatten Input

👉 Question:
- Why do we need to flatten the image for ANN?
- What will be the new shape after flattening?
- What information is lost during flattening?

#### Why flatten?
- ANN accepts 1D input, not 2D images

#### New shape:
- 28 × 28 → 784

#### Information lost:
-Spatial relationships (pixel position info)

```python
x_train_flat = x_train.reshape(-1, 28*28)
x_test_flat = x_test.reshape(-1, 28*28)
```

## Step 5: Build ANN Model

👉 Question:
- What is a Dense layer?
- Why is ReLU used in hidden layers?
- Why is Softmax used in the output layer?
- How many neurons are required in the output layer and why?

#### Dense layer:
- Fully connected layer (every neuron connects to all inputs)

#### ReLU:
- Avoids vanishing gradient, faster training

#### Softmax:
- Converts output into probabilities

#### Output neurons:
- 10 (digits 0–9)

```python
ann_model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

## Step 6: Compile ANN Model

👉 Question:
- What is an optimizer?
- Why is 'adam' commonly used?
- What is the purpose of loss function?
- Why use sparse categorical crossentropy

#### Optimizer:
- Algorithm to update weights

#### Why Adam:
- Adaptive learning rate, efficient

#### Loss function:
- Measures prediction error
- Sparse categorical crossentropy:
- Used when labels are integers (not one-hot)

```python
ann_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

## Step 7: Train ANN Model

👉 Question:
- What is an epoch?
- What happens during training?
- How does accuracy change across epochs

#### Epoch:
- One full pass through dataset
- During training:
- Weights updated via backpropagation

#### Accuracy trend:
- Increases over epochs

```python
ann_history = ann_model.fit(
    x_train_flat, y_train,
    epochs=5,
    validation_split=0.1
)
```

## Step 8: Evaluate ANN Model

👉 Question:
- What is model evaluation?
- What accuracy did you get?
- Is the model overfitting or underfitting?

#### Model evaluation:
- Testing model on unseen data

#### Accuracy:
- ~97–98%

#### Overfitting/underfitting:
- Slight overfitting possible

```python
ann_test_loss, ann_test_acc = ann_model.evaluate(x_test_flat, y_test)
print("ANN Test Accuracy:", ann_test_acc)
```

# PART 2: CNN MODEL


## Step 9: Reshape Data for CNN

👉 Question:
- Why does CNN require 4D input?
- What does the channel dimension represent?
- What will be the new shape?

#### Why 4D input:
- CNN expects (batch, height, width, channels)

#### Channel dimension:
- 1 for grayscale

#### New shape:
- (28, 28, 1)

```python
x_train_cnn = x_train.reshape(-1, 28, 28, 1)
x_test_cnn = x_test.reshape(-1, 28, 28, 1)
```

## Step 10: Build CNN Model

👉 Question:
- What is a convolution layer?
- What is a filter?
- What is pooling and why is it used?
- Why do we use Flatten after convolution?

#### Convolution layer:
- Extracts features using filters

#### Filter:
- Small matrix detecting patterns

#### Pooling:
- Reduces dimensions, avoids overfitting

#### Flatten:
- Converts feature maps to 1D

```python
cnn_model = keras.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

## Step 11: Compile CNN Model

👉 Question:
- Are compile parameters same as ANN?
- Why or why not?

#### Same parameters?
- Yes (optimizer, loss, metrics)

#### Why:
- Same classification problem

```python
cnn_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

## Step 12: Train CNN Model

👉 Question:
- Does CNN train faster or slower than ANN?
- Observe accuracy improvement

#### Training speed:
- Slower than ANN

#### Accuracy:
- Improves better than ANN

```python
cnn_history = cnn_model.fit(
    x_train_cnn, y_train,
    epochs=5,
    validation_split=0.1
)
```

## Step 13: Evaluate CNN Model

👉 Question:
- What is the CNN accuracy?
- Compare with ANN

#### CNN accuracy:
- ~99%

#### Comparison:
- Better than ANN

```python
cnn_test_loss, cnn_test_acc = cnn_model.evaluate(x_test_cnn, y_test)
print("CNN Test Accuracy:", cnn_test_acc)
```

# PART 3: COMPARISON


## Step 14: Compare Results

👉 Question:
- Which model has higher accuracy?
- Which model trained faster?
- Which model handles images better?

#### Higher accuracy:
- CNN

#### Faster training:
- ANN

#### Better for images:
- CNN

```python
print("ANN Accuracy:", ann_test_acc)
print("CNN Accuracy:", cnn_test_acc)
```
